Oh-oh, the pixel data is compressed and pydicom does not know how to handle that. This can be verified by the following.

```python
ds_anon.pixel_array
```

One way around it is through a program, `gdcmconv`, in a DICOM package **gdcm**. See <url>http://gdcm.sourceforge.net/html/gdcmconv.html</url>. 

`$ gdcmconv --raw` to decompress the DICOM pixels. gdcm is installed on cbica-cluster. You can also build it on your desktop.
